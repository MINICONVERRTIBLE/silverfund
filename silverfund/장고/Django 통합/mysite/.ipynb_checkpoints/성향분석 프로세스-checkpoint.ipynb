```python
def risk_check(funds): #위험부담수준 검색 함수
    Search_funds = list(funds)
    for x in Search_funds:
        if x not in list(df9['펀드명']):
            print(f'{x}의 위험등급정보가 존재하지 않습니다.')
        else:
            index = df9[df9['펀드명']==str(x)].index[0]
            risk=df9['위험등급'][index]
            print(f'{x}의 위험등급은 {risk}입니다.')
```
