```python
import pandas as pd
import json
import requests
```

```python
url = 'https://100lifeplan.fss.or.kr/openapi/api/rpCorpResultList.json?key=63ce3944f4cf8cc6b9f861d17c72edf9&year=2021&quarter=4&sysType=2'
res = requests.get(url)
contents = json.loads(res.text)

# with open(url, 'r') as j:
#     contents = json.loads(j.read())

contents
```

```python
contents['list'][1]['list'][0]
```

## 데이터 프레임 만들기

```python
company = []
db = []
dc = []
irp = []
irp3 = []
irp5 = []
irp7 = []
irp10 = []

for i in range(len(contents['list'])):
    company.append(contents['list'][i]['company'])
    try:
        db.append(contents['list'][i]['list'][0]['dbEarnRate'])
    except:
        db.append(None)
    try:
        dc.append(contents['list'][i]['list'][0]['dcEarnRate'])
    except:
        dc.append(None)
    try:
        irp.append(contents['list'][i]['list'][0]['irpEarnRate'])
    except:
        irp.append(None)
    try:
        irp3.append(contents['list'][i]['list'][0]['irpEarnRate3'])
    except:
        irp3.append(None)
    try:
        irp5.append(contents['list'][i]['list'][0]['irpEarnRate5'])
    except:
        irp5.append(None)
    try:
        irp7.append(contents['list'][i]['list'][0]['irpEarnRate7'])
    except:
        irp7.append(None)
    try:
        irp10.append(contents['list'][i]['list'][0]['irpEarnRate10'])
    except:
        irp10.append(None)
```

```python
df = pd.DataFrame({'company': company, 'db': db, 'dc': dc, 'irp': irp, 'irp3': irp3, 'irp5': irp5, 'irp7': irp7, 'irp10': irp10})
df
```

```python
df.fillna(0,axis=0, inplace=True)
```

## 은행, 보험회사, 증권사 수익률 비교

```python
irp_s = []
irp_i = []
irp_b = []

for idx in df.index:
    if '보험' in df.loc[idx, 'company']:
        irp_i.append(df.loc[idx, 'irp'])
    elif '은행' in df.loc[idx, 'company']:
        irp_b.append(df.loc[idx, 'irp'])
    else:
        irp_s.append(df.loc[idx, 'irp'])
```

```python
len(irp_s) + len(irp_i) + len(irp_b)
```

```python
import numpy as np

#irp단순비교
np.array(irp_s).mean(), np.array(irp_i).mean(), np.array(irp_b).mean()
```

```python
#10년 irp 수익률 비교
irp10_s = []
irp10_i = []
irp10_b = []

for idx in df.index:
    if '보험' in df.loc[idx, 'company']:
        irp10_i.append(df.loc[idx, 'irp10'])
    elif '은행' in df.loc[idx, 'company']:
        irp10_b.append(df.loc[idx, 'irp10'])
    else:
        irp10_s.append(df.loc[idx, 'irp10'])
```

```python
#10년 수익률 없는 데이터 삭제
remove_set = {0}
irp10_s= [i for i in irp10_s if i not in remove_set]
irp10_i= [i for i in irp10_i if i not in remove_set]
irp10_b = [i for i in irp10_b if i not in remove_set]
```

```python
#10년 평균 수익률 비교
np.array(irp10_s).mean(), np.array(irp10_i).mean(), np.array(irp10_b).mean()
```

## 증권사 별 수익률 비교
